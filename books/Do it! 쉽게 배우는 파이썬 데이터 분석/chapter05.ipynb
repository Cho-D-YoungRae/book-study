```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
mpg['total'].mean()
```

```python
mpg['total'].describe()
```

```python
mpg['total'].plot.hist()
```

`np.where()` 은 지정한 조건에 맞을 때와 맞지 않을 때 서로 다른 값을 반환하는 기능

```python
mpg['test'] = np.where(mpg['total'] >= 20, "pass", "fail")
mpg.head()
```

```python
mpg['test'].value_counts()
```

```python
mpg['test'].value_counts().plot.bar()
```

```python
mpg['test'].value_counts().plot.bar(rot=0)
```
