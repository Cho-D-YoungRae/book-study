```python
import pandas as pd
import numpy as np
import sys

print(f"Python version: {sys.version}")
print(f"Pandas version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
```

```python
df = pd.DataFrame({
    "sex": ["M", "F", np.nan, "M", "F"],
    "score": [5, 4, 3, 4, np.nan]
})

df
```

```python
df['score'] + 1
```

```python
pd.isna(df)
```

```python
pd.isna(df).sum()
```

```python
df.dropna(subset = ['score'])
```

```python
df.dropna(subset = ['score', 'sex'])
```

```python
df.dropna()
```

`pd.mean()`, `pd.sum()` 과 같은 수치 연산 함수는 결측치가 있으면 자동으로 제거하고 연산

`groupby()`, `agg()` 를 이용해 집단 별 요약 통계량을 구할 때도 결측치를 제거
