```python
from dotenv import load_dotenv

load_dotenv("../.env")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
```

```python
url = "https://raw.githubusercontent.com/langchain-kr/langchain-tutorial/refs/heads/main/Ch02.%20RAG/Data/2024_KB_%EB%B6%80%EB%8F%99%EC%82%B0_%EB%B3%B4%EA%B3%A0%EC%84%9C_%EC%B5%9C%EC%A2%85.pdf"
loader = PyPDFLoader(url)
pages = loader.load()

print(f"총 글자 수: {len(''.join([i.page_content for i in pages]))}")
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)

texts = text_splitter.split_documents(pages)
print('분할된 청크 수:', len(texts))
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import PyPDFLoader
```

```python
url = "https://raw.githubusercontent.com/langchain-kr/langchain-tutorial/refs/heads/main/Ch02.%20RAG/Data/2024_KB_%EB%B6%80%EB%8F%99%EC%82%B0_%EB%B3%B4%EA%B3%A0%EC%84%9C_%EC%B5%9C%EC%A2%85.pdf"
loader = PyPDFLoader(url)
pages = loader.load()

text_splitter = SemanticChunker(embeddings=OpenAIEmbeddings())

chunks = text_splitter.split_documents(pages)
```

```python
print('분할된 청크의 수:', len(chunks))
```
